```python
# !pip install selenium
# !apt-get update # to update ubuntu to correctly run apt install
# !apt install chromium-chromedriver
# !cp /usr/lib/chromium-browser/chromedriver /usr/bin
# import sys
# sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')

# from bs4 import BeautifulSoup as BS
# from selenium import webdriver
# from selenium.webdriver.chrome.options import Options
# from functools import reduce
# import pandas as pd
# import time

def render_page(url):
   
  chrome_options = Options()
  chrome_options.add_argument('--headless')
  chrome_options.add_argument('--no-sandbox')
  chrome_options.add_argument('--disable-dev-shm-usage')
  d = webdriver.Chrome('/usr/lib/chromium-browser/chromedriver',chrome_options=chrome_options)
  d.get(url) 
  time.sleep(3)
  r = d.page_source
  d.quit()
 
  return r

# print(render_page("https://www.wunderground.com/history/daily/eg/al-shorouk-city/HECA/date/2022-2-19"))



def my_scrap(dates):
  
  
  for d in dates:


    url = "https://www.wunderground.com/history/daily/eg/al-khalifa/HECA/date/"+d
    s=url


    

    r = render_page(url)
        

    soup = BS(r, "html.parser")
    container = soup.find('lib-city-history-observation')
    check = container.find('tbody')
    
    data = []

    try:
        for c in check.find_all('tr', class_='ng-star-inserted'):


          for i in c.find_all('td', class_='ng-star-inserted'):

            trial = i.text
            trial = trial.strip('  ')
            trial = trial.strip('AM')
            trial = trial.strip('PM')
            trial = trial.strip('\xa0°F')
            trial = trial.strip('\xa0°%')
            trial = trial.strip('\xa0°mph')
            trial = trial.strip('\xa0°in')
            data.append(trial)
                        
                        
          Time=pd.DataFrame([data[0:len(data)][x] for x in range(0,len(data),10)],columns=["Time"])
          Temperature=pd.DataFrame([data[0:len(data)][x+1] for x in range(0,len(data),10)],columns=["Temperature"])
          Dew_Point=pd.DataFrame([data[0:len(data)][x+2] for x in range(0,len(data),10)],columns=["Dew Point"])
          Humidity=pd.DataFrame([data[0:len(data)][x+3] for x in range(0,len(data),10)],columns=["Humidity"])
          Wind=pd.DataFrame([data[0:len(data)][x+4] for x in range(0,len(data),10)],columns=["Wind"])
          Wind_Speed=pd.DataFrame([data[0:len(data)][x+5] for x in range(0,len(data),10)],columns=["Wind Speed"])
          Wind_Gust=pd.DataFrame([data[0:len(data)][x+6] for x in range(0,len(data),10)],columns=["Wind Gust"])
          Pressure=pd.DataFrame([data[0:len(data)][x+7] for x in range(0,len(data),10)],columns=["Pressure"])
          Precip=pd.DataFrame([data[0:len(data)][x+8] for x in range(0,len(data),10)],columns=["Precip"])
          Condition=pd.DataFrame([data[0:len(data)][x+9] for x in range(0,len(data),10)],columns=["Condition"])
            
          dfs = [Time,Temperature,Dew_Point,Humidity,Wind,Wind_Speed,Wind_Gust,Pressure,Precip,Condition]
          df_final = reduce(lambda left, right: pd.merge(left, right, left_index=True, right_index=True), dfs)
          df_final["Date"]=d
            
          global final  
          final = final.append(df_final)
    except:
        pass
        
        
  return final

                    
        
import datetime
import pandas as pd
 



# initializing date
test_date = datetime.datetime.strptime("2022-11-28", "%Y-%m-%d")
 
# initializing K
K = 365
 
date_generated = pd.date_range(test_date, periods=100)
s=list(date_generated.strftime("%Y-%m-%d"))


ff=my_scrap(s)
```
